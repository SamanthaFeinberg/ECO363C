### Notebook Basics
Having launched Jupyter Notebook and opened a new notebook, you should now see an _active cell_, into which you can type Python commands. Notice that the cell is surrounded by a green or blue border, which means that the cell is in _edit mode_ (Jupyter Notebook recently changed the color of the border of the cell in edit mode. Independent of the border color, what definitely indicates that you are in edit mode is a _blinking cursor_ in the cell). That means, you can type in new commands and comments simply by typing.

When you want to execute the written code in a given cell, hit either 

 - ```Shift-Enter``` which executes the code in the cell and directly opens a new cell to continue coding, or
 - ```Cmd-Enter```which executes the code in the cell without opening a new cell.
 
If you simply hit ```Enter```, you will get a new line within a cell and the current cell is not executed.

Instead of using these keyboard shortcuts, there are also menu and botton options for running code in a cell.

Hint: If you want to learn more about the basics of these notebooks, in particular about keyboard shortcuts, click on `Help` at the top.

### Modal Editing
Jupyter notebook uses a _modal editing_ system. That means that the effect of typing at the keyboard __depends on which mode you are in.__

There are two modes:

 1. Edit mode:
     - indicated by a blinking cursor
     - whatever you type appears as is in that cell
     
 2. Command mode:
     - there is no blinking cursor
     - Keystrokes are interpreted as commands. For example, if you type ```b``` it creates a new cell below the current one.
     
You can switch between the two:

 - to go from command mode to edit mode, hit ```Enter```or click in a cell
 - to go from edit mode to command mode, hit the ```Esc``` key or ```Ctrl-M```.

### Inserting Unicode 
Python supports [unicode](https://docs.python.org/3/howto/unicode.html), which allows the use of characters such as $\alpha$ and $\beta$ as names in your code.

To try this, go into a code cell, type ```\alpha```and then hit the tab key on your keyboard (we will see this at work in the following code, with `\theta` not `\alpha`).

Let's run a test program.

````python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(78712)

# Compute pie slices
N = 20
θ = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(projection='polar')
ax.bar(θ, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

plt.show()
````

Here is an example of using LaTeX code in Markdown:

A household's utility function $U(\cdot)$ depends on the household's consumption $C_t$ and labor hours worked $N_t$ as follows:
$$U(C_t,N_t) = \begin{cases}
                \frac{C_t^{1-\sigma} - 1}{1-\sigma} - \psi\frac{N_t^{1+\nu}}{1+\nu}, \text{ if $\sigma \neq 1$}\\
                \log(C_t) - \psi\frac{N_t^{1+\nu}}{1+\nu}, \text{ if $\sigma = 1$}
                \end{cases}
                $$

where $\sigma > 0$ denotes the household's risk aversion, $\nu > 0$ the inverse Frisch elasticity, and $\psi > 0$ the disutility weight on hours worked.

### Working with Python files
Executing Python code entered in a Jupyter notebook cell has become "the norm", but traditionally this was not the case. Instead, code used to be saved first in a text file on a local machine (with the ```.py```extension).

We can easily create such a file as follows:

````python
%%writefile firstfile.py
print("Hello, this is my first Python file")
````

This program writes the line ```print("Hello, this is my first Python file")```into a file called ```firstfile.py```in the local directory. 

Here ```%%writefile```is an example of a [cell magic](https://ipython.readthedocs.io/en/stable/interactive/magics.html#cell-magics).
